```python
from tools import Tools
from helpers import load_all_data, load_window
```

```python
# Load market cap and debt data
market_cap_data, debt = load_all_data()
ticker_list = ['SAP GY Equity',	
               'MRK GY Equity', 
               'AI FP Equity', 
               'CRH LN Equity', 
               'SRG IM Equity', 
               'DAI GY Equity', 
               'VIE FP Equity', 
               'AMP IM Equity', 
               'FR FP Equity', 
               'EO FP Equity', 
               'GET FP Equity', 
               'LHA GY Equity', 
               'PIA IM Equity', 
               'CO FP Equity']
```

# 1Y non stressed [2017-10-13 / 2018-10-01]

```python
# Define date range and load market capitalization data
start_date = '2017-10-13'
end_date = '2018-10-01'
market_cap = load_window(market_cap_data, start_date, end_date)

# Initialize Tools class and perform calibrations
helper_1Y_non_stressed = Tools(market_cap, debt, ticker_list)
helper_1Y_non_stressed.calibrate_and_combine_results(ticker_list)

# Save results and retrieve the combined results DataFrame
helper_1Y_non_stressed.dump_results("1Y", "non_stressed")
res_1y_non_stressed = helper_1Y_non_stressed.get_results()

#display results
res_1y_non_stressed
```

# 1Y stressed [2019/10/28 - 2020/10/13]

```python
# Define date range and load market capitalization data
start_date = '2019-10-28'
end_date = '2020-10-13'
market_cap = market_cap_data.set_index("Dates").loc[start_date:end_date]

# Initialize Tools class and perform calibrations
helper_1Y_stressed = Tools(market_cap, debt, ticker_list)
helper_1Y_stressed.calibrate_and_combine_results(ticker_list)

# Save results and retrieve the combined results DataFrame
helper_1Y_stressed.dump_results("1Y", "stressed")

#display results
helper_1Y_stressed.get_results()
```
